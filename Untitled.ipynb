```python
import tweepy
import requests
import json
import matplotlib.pyplot as plt
from numpy import mean
import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
tweets = api.user_timeline('@nytimes')
```

```python
analyzer.polarity_scores(api.user_timeline('@nytimes')[0]['text'])
```

```python
num_tweets = 40
max_page = round(num_tweets/20)
result_dict = {}
users = ['@NYTimes', '@BBC', '@FOX', '@CBS', '@CNN']
for user in users:
    result_dict[user] = {}
    result_dict[user]['compound'] = []
    result_dict[user]['pos'] = []
    result_dict[user]['neg'] = []
    result_dict[user]['neu'] = []
    result_dict[user]['text'] = []
    result_dict[user]['time'] = []
    
    if True:
        for x in range(max_page):
            tweets = api.user_timeline(user,page = x)
            for tweet in tweets:
                poles = analyzer.polarity_scores(tweet['text'])
                result_dict[user]['compound'].append(poles['compound'])
                result_dict[user]['pos'].append(poles['pos'])
                result_dict[user]['neg'].append(poles['neg'])
                result_dict[user]['neu'].append(poles['neu'])
                result_dict[user]['text'].append(tweet['text'])
                result_dict[user]['time'].append(tweet['created_at'])
print(json.dumps(tweet, indent = 2, sort_keys=True))
```

```python
df_dict = {}
for user in users:
    df_dict[user] = pd.DataFrame(result_dict[user])
df_dict[user].head()
```

```python
scatterdict = {}
for user in users:
    scatterdict[user] = df_dict[user]['compound']

pd.DataFrame(scatterdict).plot(marker = 'o', linewidth = 0)
plt.xlabel('Tweets Ago')
plt.ylabel('Sentiment')
plt.title(f'plot of sentiment at {df_dict[users[0]]["time"][0]}')
plt.show()
```
